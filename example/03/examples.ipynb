```python
# Tkinteræº«åº¦è½‰æ›å™¨å®Œæ•´ç‰ˆ
import tkinter as tk
from tkinter import ttk

class TemperatureConverter:
    def __init__(self, root):
        self.root = root
        self.root.title("æº«åº¦è½‰æ›å™¨")
        self.root.geometry("300x200")

        # è¼¸å…¥æ¡†
        ttk.Label(root, text="æº«åº¦:").pack(pady=5)
        self.temp_entry = ttk.Entry(root)
        self.temp_entry.pack(pady=5)

        # æŒ‰éˆ•
        ttk.Button(root, text="æ”æ°â†’è¯æ°",
                  command=self.c_to_f).pack(pady=5)
        ttk.Button(root, text="è¯æ°â†’æ”æ°",
                  command=self.f_to_c).pack(pady=5)

        # çµæœé¡¯ç¤º
        self.result_label = ttk.Label(root, text="çµæœå°‡é¡¯ç¤ºåœ¨é€™è£¡")
        self.result_label.pack(pady=10)

    def c_to_f(self):
        try:
            celsius = float(self.temp_entry.get())
            fahrenheit = celsius * 9/5 + 32
            self.result_label.config(
                text=f"{celsius}Â°C = {fahrenheit:.1f}Â°F"
            )
        except ValueError:
            self.result_label.config(text="è«‹è¼¸å…¥æœ‰æ•ˆæ•¸å­—ï¼")

    def f_to_c(self):
        try:
            fahrenheit = float(self.temp_entry.get())
            celsius = (fahrenheit - 32) * 5/9
            self.result_label.config(
                text=f"{fahrenheit}Â°F = {celsius:.1f}Â°C"
            )
        except ValueError:
            self.result_label.config(text="è«‹è¼¸å…¥æœ‰æ•ˆæ•¸å­—ï¼")

# åŸ·è¡Œç¨‹å¼
# if __name__ == "__main__":
#     root = tk.Tk()
#     app = TemperatureConverter(root)
#     root.mainloop()

print("Tkinter æº«åº¦è½‰æ›å™¨é¡åˆ¥å·²å®šç¾©å®Œæˆ")
print("è‹¥è¦åŸ·è¡ŒGUIï¼Œè«‹å–æ¶ˆè¨»è§£æœ€å¾Œä¸‰è¡Œç¨‹å¼ç¢¼")
```
