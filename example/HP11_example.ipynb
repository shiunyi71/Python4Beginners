## ÈÄ≤ÈöéÂÖßÂª∫ÂáΩÂºè

```python
# enumerate() - ÂèñÂæóÁ¥¢ÂºïÂíåÂÄº
fruits = ['ËòãÊûú', 'È¶ôËïâ', 'Ê©òÂ≠ê']
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
# Ëº∏Âá∫Ôºö
# 0: ËòãÊûú
# 1: È¶ôËïâ
# 2: Ê©òÂ≠ê

# zip() - ÈÖçÂ∞çÂ§öÂÄãÂ∫èÂàó
names = ['Â∞èÊòé', 'Â∞èËèØ', 'Â∞èÁæé']
scores = [85, 92, 78]
print(tuple(zip(names, scores)))
for name, score in zip(names, scores):
    print(f"{name}: {score}ÂàÜ")

# map() - Â∞çÊØèÂÄãÂÖÉÁ¥†Â•óÁî®ÂáΩÂºè
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)  # [1, 4, 9, 16, 25]

# filter() - ÈÅéÊøæÂÖÉÁ¥†
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # [2, 4]
```

## üîß Ëá™ÂÆöÁæ©Ê®°ÁµÑÂØ¶Êà∞ - magic_tools.py

ÈÄôÊòØ‰∏ÄÂÄãËá™ÂÆöÁæ©Ê®°ÁµÑÁöÑÁØÑ‰æãÔºåÈúÄË¶ÅÂè¶Â§ñÂª∫Á´ãÊ™îÊ°à„ÄÇ

```python
# filepath: magic_tools.py
"""
È≠îÊ≥ïÂ∑•ÂÖ∑Ê®°ÁµÑ
Êèê‰æõÂêÑÁ®ÆÂØ¶Áî®ÁöÑÈ≠îÊ≥ïÂáΩÂºè
"""

def fire_ball(target, power=30):
    """ÁÅ´ÁêÉË°ì"""
    damage = power * 1.5
    return f"üî• Â∞ç{target}ÈÄ†Êàê{damage}ÈªûÁÅ´ÁÑ∞ÂÇ∑ÂÆ≥ÔºÅ"

def heal(target, amount=50):
    """Ê≤ªÁôÇË°ì"""
    return f"‚ú® ÁÇ∫{target}ÊÅ¢Âæ©{amount}ÈªûHPÔºÅ"

def calculate_damage(base, multiplier=1.0):
    """ÂÇ∑ÂÆ≥Ë®àÁÆó"""
    return int(base * multiplier)

# Ê®°ÁµÑÁ¥öÂà•ÁöÑËÆäÊï∏
GAME_VERSION = "1.0.0"
MAX_LEVEL = 99

# Ê∏¨Ë©¶Á®ãÂºèÁ¢ºÔºàÂè™ÊúâÁõ¥Êé•Âü∑Ë°åÊ≠§Ê™îÊ°àÊâçÊúÉÂü∑Ë°åÔºâ
if __name__ == "__main__":
    print("Ê∏¨Ë©¶È≠îÊ≥ïÂ∑•ÂÖ∑Ê®°ÁµÑ...")
    print(fire_ball("Âì•Â∏ÉÊûó", 50))
    print(heal("ÂãáËÄÖ", 80))
```

## character_tools.py - ËßíËâ≤ÁÆ°ÁêÜÂ∑•ÂÖ∑Ê®°ÁµÑ

```python
# filepath: character_tools.py
"""ËßíËâ≤ÁÆ°ÁêÜÂ∑•ÂÖ∑"""

def create_character(name, job, level=1):
    """ÂâµÂª∫ËßíËâ≤"""
    return {
        "name": name,
        "job": job,
        "level": level,
        "hp": 100 * level,
        "mp": 50 * level
    }

def level_up(character):
    """ËßíËâ≤ÂçáÁ¥ö"""
    character["level"] += 1
    character["hp"] = 100 * character["level"]
    character["mp"] = 50 * character["level"]
    print(f"üéâ {character['name']} ÂçáÂà∞ Lv.{character['level']}ÔºÅ")
    return character

def show_status(character):
    """È°ØÁ§∫ËßíËâ≤ÁãÄÊÖã"""
    print(f"\n{'='*40}")
    print(f"‚öîÔ∏è  ËßíËâ≤Ôºö{character['name']}")
    print(f"üé≠ ËÅ∑Ê•≠Ôºö{character['job']}")
    print(f"‚≠ê Á≠âÁ¥öÔºöLv.{character['level']}")
    print(f"‚ù§Ô∏è  HPÔºö{character['hp']}")
    print(f"üíô MPÔºö{character['mp']}")
    print(f"{'='*40}\n")
```

```python
# filepath: main.py
"""‰∏ªÁ®ãÂºèÔºöÊï¥ÂêàÊâÄÊúâÊ®°ÁµÑ"""

# import magic_tools
# import character_tools

# ÂâµÂª∫ËßíËâ≤
hero = create_character("‰∫ûÁëü", "È®éÂ£´", 5)
show_status(hero)

# ‰ΩøÁî®È≠îÊ≥ï
print(fire_ball("ÊöóÈªëÈ®éÂ£´", 60))
print(heal(hero["name"], 50))

# ËßíËâ≤ÂçáÁ¥ö
hero = level_up(hero)
show_status(hero)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ÂâµÂª∫2x2ÁöÑÂ≠êÂúñ
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('RPGÈÅäÊà≤Êï∏ÊìöÂàÜÊûê üìä', fontsize=18)

# Â≠êÂúñ1ÔºöÁ∂ìÈ©óÂÄºÊõ≤Á∑ö
x = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.exp(x/3), 'b-', linewidth=2)
axes[0, 0].set_title('Á∂ìÈ©óÂÄºÊàêÈï∑Êõ≤Á∑ö')
axes[0, 0].set_xlabel('Á≠âÁ¥ö')
axes[0, 0].set_ylabel('ÊâÄÈúÄÁ∂ìÈ©óÂÄº')
axes[0, 0].grid(True, alpha=0.3)

# Â≠êÂúñ2ÔºöËßíËâ≤Â±¨ÊÄß
axes[0, 1].bar(['HP', 'MP', 'ATK', 'DEF'], [100, 80, 90, 85],
              color=['red', 'blue', 'orange', 'green'])
axes[0, 1].set_title('ËßíËâ≤Â±¨ÊÄß')
axes[0, 1].set_ylabel('Êï∏ÂÄº')

# Â≠êÂúñ3ÔºöÊïµ‰∫∫ÂàÜÂ∏É
axes[1, 0].scatter(np.random.rand(50), np.random.rand(50),
                  s=100, alpha=0.6, c='red')
axes[1, 0].set_title('Êïµ‰∫∫ÂàÜÂ∏É')
axes[1, 0].set_xlabel('XÂ∫ßÊ®ô')
axes[1, 0].set_ylabel('YÂ∫ßÊ®ô')

# Â≠êÂúñ4ÔºöÂÖÉÁ¥†ÂàÜÂ∏É
axes[1, 1].pie([30, 25, 25, 20],
              labels=['ÁÅ´üî•', 'Ê∞¥üíß', 'È¢®üå™Ô∏è', 'Âúüüå±'],
              autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('ÂÖÉÁ¥†ÂàÜÂ∏É')

plt.tight_layout()
plt.show()
```

## ÈÄ≤ÈöéÊåëÊà∞ÔºöÊ®ôË®ªÊúÄÈ´òÂÄº

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hours = [2, 3, 2.5, 4, 3, 5, 4]

plt.figure(figsize=(10, 6))
plt.plot(days, hours, marker='o', linewidth=2, markersize=8)

# Ê®ôË®ªÊúÄÈ´òÂÄº
max_hours = max(hours)
max_day = days[hours.index(max_hours)]
plt.annotate(f'ÊúÄÈ´ò: {max_hours}Â∞èÊôÇ',
            xy=(max_day, max_hours),
            xytext=(10, 10),
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='red'))

plt.title('ÊØèÈÄ±Â≠∏ÁøíÊôÇÊï∏Áµ±Ë®à')
plt.xlabel('ÊòüÊúü')
plt.ylabel('ÊôÇÊï∏')
plt.grid(True, alpha=0.3)

# ÂÑ≤Â≠òÂúñË°®
plt.savefig('study_hours.png', dpi=300, bbox_inches='tight')
plt.show()
```
