## 魔法傷害計算

```python
def calculate_damage(base_damage, critical=False):
    """計算攻擊傷害"""
    if critical:
        damage = base_damage * 2
        print("💥 暴擊2x！")
    else:
        damage = base_damage

    return damage

# 使用函式
normal_hit = calculate_damage(50)
print(f"普通攻擊傷害：{normal_hit}")

critical_hit = calculate_damage(50, critical=True)
print(f"暴擊傷害：{critical_hit}")
```

## 回傳多個值

```python
def battle_result(attacker_hp, damage_taken):
    """計算戰鬥結果"""
    remaining_hp = attacker_hp - damage_taken
    if remaining_hp < 0:
      remaining_hp = 0
    is_alive = remaining_hp > 0

    return remaining_hp, is_alive  # 回傳多個值

# 接收多個回傳值
hp, alive = battle_result(100, 60)
print(f"剩餘HP：{hp}")
print(f"是否存活：{alive}")

hp, alive = battle_result(hp, 401)
print(f"剩餘HP：{hp}")
print(f"是否存活：{alive}")
```

```python
player_hp = 100  # 全域變數

def take_damage(damage):
    global player_hp  # 宣告要使用全域變數
    player_hp = player_hp - damage
    print(f"💔 受到 {damage} 點傷害！")
    print(f"❤️ 剩餘HP：{player_hp}")

take_damage(30)
print(f"當前HP：{player_hp}")  # HP真的改變了
```

## 範例2：治療魔法

```python
def heal(target, heal_amount=50):
    """治療術"""
    print(f"✨ 對{target}施放治療術！")
    print(f"💚 恢復{heal_amount}點HP！")
    return heal_amount

def check_hp(current_hp, max_hp=100):
    """檢查HP狀態"""
    percentage = (current_hp / max_hp) * 100

    if percentage <= 30:
        status = "⚠️ 危險"
    elif percentage <= 60:
        status = "⚡ 注意"
    else:
        status = "✅ 健康"

    return status

# 使用魔法
heal("勇者", 40)
status = check_hp(70)
print(f"狀態：{status}")

hp_player1 = 10
status = check_hp(hp_player1)
print(f"小明 狀態：{status}")
hp_player1 = heal("小明",hp_player1+40)
print(f"小明 剩餘HP：{hp_player1}")
status = check_hp(hp_player1)
print(f"小明 狀態：{status}")
```

```python
def combo_attack(character, *targets, power=50):
    """
    對多個目標發動連擊
    *targets 可以接收任意數量的參數
    """
    print(f"⚔️ {character} 發動連擊技！")
    total_damage = 0

    for target in targets:
        damage = power
        print(f"  💥 攻擊 {target}，造成 {damage} 傷害")
        total_damage += damage

    print(f"✨ 總傷害：{total_damage}")
    return total_damage

# 攻擊多個敵人
combo_attack("劍士", "哥布林A", "哥布林B", "哥布林C", power=40)
```

## 函式設計原則：好的設計

```python
# ✅ 好的設計：功能單一、名稱清楚
def calculate_total_price(price, quantity):
    """計算總價"""
    return price * quantity

def apply_discount(price, discount_rate):
    """套用折扣"""
    return price * (1 - discount_rate)

# 使用
total = calculate_total_price(100, 3)
final_price = apply_discount(total, 0.1)
print(f"最終價格：{final_price}")
```

```python
# 1. 創建角色
def create_hero(name, job, hp=100, mp=50, lv=1):
    """創建英雄角色"""
    dc = {'name':name,'job':job,'hp':hp,'mp':mp,'lv':lv}
    return dc

# 2. 升級角色
def level_up(hero, up=1):
    """角色升級"""
    hero['lv'] += up

# 3. 顯示角色資訊
def show_status(hero):
    for key,value in hero.items():
        print(key,':',value)

hero1 = create_hero("亞瑟", "騎士")
level_up(hero1)
show_status(hero1)

hero2 = create_hero("梅林", "法師")
level_up(hero2, 10)
show_status(hero2)
```
