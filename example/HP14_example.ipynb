## ç¯„ä¾‹ 5: å®Œæ•´çš„å¯¶å¯å¤¢é¡åˆ¥

```python
# filepath: example_05_complete_pokemon.py
class Pokemon:
    def __init__(self, name, type, level, hp, attack_power):
        self.name = name
        self.type = type            # å±¬æ€§ï¼ˆé›»ã€æ°´ã€ç«ç­‰ï¼‰
        self.level = level
        self.hp = hp
        self.max_hp = hp
        self.attack_power = attack_power

    def show_status(self):
        """é¡¯ç¤ºç‹€æ…‹"""
        hp_bar = "â¤ï¸" * (self.hp // 10) + "ğŸ–¤" * ((self.max_hp - self.hp) // 10)
        print(f"\n{'='*40}")
        print(f"ğŸ¯ åå­—ï¼š{self.name} ({self.type}ç³»)")
        print(f"   ç­‰ç´šï¼šLv.{self.level}")
        print(f"   HPï¼š{hp_bar} {self.hp}/{self.max_hp}")
        print(f"   æ”»æ“ŠåŠ›ï¼š{self.attack_power}")
        print(f"{'='*40}")

    def attack(self, target):
        """æ”»æ“Šç›®æ¨™"""
        if self.hp <= 0:
            print(f"âŒ {self.name} å·²ç¶“ç„¡æ³•æˆ°é¬¥äº†ï¼")
            return

        damage = self.attack_power + self.level * 2
        target.hp = max(0, target.hp - damage)

        print(f"\nâš”ï¸ {self.name} ä½¿ç”¨æ”»æ“Šï¼")
        print(f"ğŸ’¥ å° {target.name} é€ æˆ {damage} é»å‚·å®³ï¼")

        if target.hp <= 0:
            print(f"ğŸ’€ {target.name} å¤±å»æˆ°é¬¥èƒ½åŠ›äº†ï¼")
            target.hp = 0
        # else:
        #     target.show_status()

    def heal(self, amount):
        """å›å¾©ç”Ÿå‘½å€¼"""
        old_hp = self.hp
        self.hp = min(self.max_hp, self.hp + amount)
        healed = self.hp - old_hp
        print(f"ğŸ’š {self.name} å›å¾©äº† {healed} é» HPï¼")
```

## ç¯„ä¾‹ 7: é€²éšå¯¶å¯å¤¢é¡åˆ¥è¨­è¨ˆ

```python
# filepath: example_07_advanced_pokemon.py
class AdvancedPokemon:
    # é¡åˆ¥å±¬æ€§ï¼šè¨˜éŒ„æ‰€æœ‰å‰µå»ºçš„å¯¶å¯å¤¢æ•¸é‡
    total_count = 0

    # é¡åˆ¥å±¬æ€§ï¼šå±¬æ€§ç›¸å‰‹è¡¨
    type_chart = {
        "ç«": {"æ°´": 0.5, "è‰": 2.0, "ç«": 0.5},
        "æ°´": {"ç«": 2.0, "è‰": 0.5, "æ°´": 0.5},
        "è‰": {"æ°´": 2.0, "ç«": 0.5, "è‰": 0.5},
        "é›»": {"æ°´": 2.0, "è‰": 0.5, "é£›è¡Œ": 2.0},
    }

    def __init__(self, name, poke_type, level=5):
        self.name = name
        self.type = poke_type
        self.level = level
        self.hp = 50 + level * 5
        self.max_hp = self.hp
        self.attack_power = 20 + level * 2
        self.exp = 0
        self.exp_to_next_level = 100

        AdvancedPokemon.total_count += 1

    def attack(self, target):
        """è€ƒæ…®å±¬æ€§ç›¸å‰‹çš„æ”»æ“Š"""
        if self.hp <= 0:
            print(f"âŒ {self.name} å·²ç¶“ç„¡æ³•æˆ°é¬¥äº†ï¼")
            return

        # è¨ˆç®—åŸºç¤å‚·å®³
        base_damage = self.attack_power + self.level * 2

        # è¨ˆç®—å±¬æ€§ç›¸å‰‹åŠ æˆ
        multiplier = self.get_type_multiplier(self.type, target.type)
        damage = int(base_damage * multiplier)

        target.hp = max(0, target.hp - damage)

        # é¡¯ç¤ºæ•ˆæœ
        if multiplier > 1:
            print(f"âš¡ æ•ˆæœçµ•ä½³ï¼")
        elif multiplier < 1:
            print(f"ğŸ›¡ï¸ æ•ˆæœä¸ä½³...")

        print(f"ğŸ’¥ {self.name} å° {target.name} é€ æˆ {damage} é»å‚·å®³ï¼")

        # ç²å¾—ç¶“é©—å€¼
        if target.hp <= 0:
            self.gain_exp(50)

    @staticmethod
    def get_type_multiplier(attacker_type, defender_type):
        """éœæ…‹æ–¹æ³•ï¼šè¨ˆç®—å±¬æ€§ç›¸å‰‹å€ç‡"""
        if attacker_type in AdvancedPokemon.type_chart:
            if defender_type in AdvancedPokemon.type_chart[attacker_type]:
                return AdvancedPokemon.type_chart[attacker_type][defender_type]
        return 1.0

    def gain_exp(self, amount):
        """ç²å¾—ç¶“é©—å€¼"""
        self.exp += amount
        print(f"â­ {self.name} ç²å¾—äº† {amount} é»ç¶“é©—å€¼ï¼")

        # æª¢æŸ¥æ˜¯å¦å‡ç´š
        while self.exp >= self.exp_to_next_level:
            self.level_up()

    def level_up(self):
        """å‡ç´š"""
        self.exp -= self.exp_to_next_level
        self.level += 1
        self.max_hp += 10
        self.hp = self.max_hp
        self.attack_power += 5
        self.exp_to_next_level = int(self.exp_to_next_level * 1.2)

        print(f"ğŸŠ æ­å–œï¼{self.name} å‡ç´šåˆ° Lv.{self.level}ï¼")

    @classmethod
    def get_total_count(cls):
        """é¡åˆ¥æ–¹æ³•ï¼šç²å–å‰µå»ºçš„å¯¶å¯å¤¢ç¸½æ•¸"""
        return cls.total_count

    def __str__(self):
        """å­—ä¸²è¡¨ç¤ºæ³•"""
        return f"{self.name}(Lv.{self.level}, HP:{self.hp}/{self.max_hp})"

    def __repr__(self):
        """é–‹ç™¼è€…å‹å–„çš„è¡¨ç¤ºæ³•"""
        return f"AdvancedPokemon('{self.name}', '{self.type}', {self.level})"
```

```python
# ä½¿ç”¨ AdvancedPokemon é¡åˆ¥å‰µå»ºä¸‰éš»å¯¶å¯å¤¢
fire_pokemon = AdvancedPokemon("å°ç«é¾", "ç«", 7)
water_pokemon = AdvancedPokemon("å‚‘å°¼é¾œ", "æ°´", 8)
grass_pokemon = AdvancedPokemon("å¦™è›™ç¨®å­", "è‰", 6)

# é¡¯ç¤ºä¸‰éš»å¯¶å¯å¤¢çš„è³‡è¨Š
print(fire_pokemon)
print(water_pokemon)
print(grass_pokemon)

# é€²è¡Œä¸€æ¬¡æ”»æ“Šä¸¦é¡¯ç¤ºçµæœ
fire_pokemon.attack(water_pokemon)
water_pokemon.attack(grass_pokemon)
grass_pokemon.attack(fire_pokemon)

# é¡¯ç¤ºæ”»æ“Šå¾Œçš„ç‹€æ…‹
print(fire_pokemon)
print(water_pokemon)
print(grass_pokemon)

# é¡¯ç¤ºç›®å‰å‰µå»ºçš„å¯¶å¯å¤¢ç¸½æ•¸
print(f"ç›®å‰å…±å‰µå»ºäº† {AdvancedPokemon.get_total_count()} éš»å¯¶å¯å¤¢")
```

## ç¯„ä¾‹ 8: å°æˆ°ç³»çµ±

```python
# filepath: example_08_battle_system.py
class BattleSystem:
    """å¯¶å¯å¤¢å°æˆ°ç³»çµ±"""

    def __init__(self, pokemon1, pokemon2):
        self.pokemon1 = pokemon1
        self.pokemon2 = pokemon2
        self.turn = 1

    def start_battle(self):
        """é–‹å§‹å°æˆ°"""
        print(f"\n{'='*50}")
        print(f"ğŸ® å°æˆ°é–‹å§‹ï¼")
        print(f"   {self.pokemon1.name} vs {self.pokemon2.name}")
        print(f"{'='*50}\n")

        while self.pokemon1.hp > 0 and self.pokemon2.hp > 0:
            self.play_turn()

        self.announce_winner()

    def play_turn(self):
        """åŸ·è¡Œä¸€å€‹å›åˆ"""
        print(f"\n--- ç¬¬ {self.turn} å›åˆ ---")

        # å¯¶å¯å¤¢1æ”»æ“Š
        if self.pokemon1.hp > 0:
            self.pokemon1.attack(self.pokemon2)

        # å¯¶å¯å¤¢2åæ“Š
        if self.pokemon2.hp > 0:
            self.pokemon2.attack(self.pokemon1)

        # é¡¯ç¤ºé›™æ–¹ç‹€æ…‹
        print(f"\nğŸ“Š ç•¶å‰ç‹€æ…‹ï¼š")
        print(f"   {self.pokemon1}")
        print(f"   {self.pokemon2}")

        self.turn += 1

        # æš«åœä¸€ä¸‹è®“ç©å®¶çœ‹æ¸…æ¥š
        input("\næŒ‰ Enter ç¹¼çºŒ...")

    def announce_winner(self):
        """å®£å¸ƒç²å‹è€…"""
        print(f"\n{'='*50}")
        if self.pokemon1.hp > 0:
            print(f"ğŸ† {self.pokemon1.name} ç²å‹ï¼")
        else:
            print(f"ğŸ† {self.pokemon2.name} ç²å‹ï¼")
        print(f"{'='*50}")
```

```python
fire_pokemon = AdvancedPokemon("å°ç«é¾", "ç«", 17)
water_pokemon = AdvancedPokemon("å‚‘å°¼é¾œ", "æ°´", 8)

# ä½¿ç”¨ BattleSystem é€²è¡Œå°æˆ°ç¤ºç¯„
battle = BattleSystem(fire_pokemon, water_pokemon)
battle.start_battle()
```

## ç¯„ä¾‹ 9: å¯¶å¯å¤¢åœ–é‘‘ç³»çµ±

```python
# filepath: example_09_pokedex.py
class Pokedex:
    """å¯¶å¯å¤¢åœ–é‘‘ç³»çµ±"""

    def __init__(self):
        self.pokemons = []  # æ”¶é›†åˆ°çš„å¯¶å¯å¤¢åˆ—è¡¨

    def add_pokemon(self, pokemon):
        """æ–°å¢å¯¶å¯å¤¢åˆ°åœ–é‘‘"""
        self.pokemons.append(pokemon)
        print(f"âœ… {pokemon.name} å·²åŠ å…¥åœ–é‘‘ï¼")
        print(f"ğŸ“š ç›®å‰æ”¶é›†äº† {len(self.pokemons)} éš»å¯¶å¯å¤¢")

    def show_all(self):
        """é¡¯ç¤ºæ‰€æœ‰æ”¶é›†çš„å¯¶å¯å¤¢"""
        if not self.pokemons:
            print("ğŸ“­ åœ–é‘‘æ˜¯ç©ºçš„ï¼Œå¿«å»æ”¶é›†å¯¶å¯å¤¢å§ï¼")
            return

        print(f"\n{'='*60}")
        print(f"ğŸ“– å¯¶å¯å¤¢åœ–é‘‘ï¼ˆå…± {len(self.pokemons)} éš»ï¼‰")
        print(f"{'='*60}")

        for i, pokemon in enumerate(self.pokemons, 1):
            print(f"{i:3d}. {pokemon}")

    def search_by_name(self, name):
        """æ ¹æ“šåå­—æœå°‹å¯¶å¯å¤¢"""
        results = [p for p in self.pokemons if name in p.name]

        if results:
            print(f"\nğŸ” æ‰¾åˆ° {len(results)} éš»å¯¶å¯å¤¢ï¼š")
            for pokemon in results:
                print(f"   â€¢ {pokemon}")
        else:
            print(f"âŒ æ‰¾ä¸åˆ°åå­—åŒ…å« '{name}' çš„å¯¶å¯å¤¢")

    def search_by_type(self, poke_type):
        """æ ¹æ“šå±¬æ€§æœå°‹å¯¶å¯å¤¢"""
        results = [p for p in self.pokemons if p.type == poke_type]

        if results:
            print(f"\nğŸ” æ‰¾åˆ° {len(results)} éš» {poke_type}ç³» å¯¶å¯å¤¢ï¼š")
            for pokemon in results:
                print(f"   â€¢ {pokemon}")
        else:
            print(f"âŒ æ‰¾ä¸åˆ° {poke_type}ç³» çš„å¯¶å¯å¤¢")

    def get_statistics(self):
        """é¡¯ç¤ºçµ±è¨ˆè³‡è¨Š"""
        if not self.pokemons:
            return

        total = len(self.pokemons)
        avg_level = sum(p.level for p in self.pokemons) / total
        max_level_pokemon = max(self.pokemons, key=lambda p: p.level)

        print(f"\nğŸ“Š åœ–é‘‘çµ±è¨ˆï¼š")
        print(f"   ç¸½æ•¸é‡ï¼š{total} éš»")
        print(f"   å¹³å‡ç­‰ç´šï¼š{avg_level:.1f}")
        print(f"   æœ€é«˜ç­‰ç´šï¼š{max_level_pokemon.name} (Lv.{max_level_pokemon.level})")
```
