## 🔄 迴圈複習

### While 迴圈 - 基本結構

prompt: 寫一個 while 迴圈，從 0 執行到 4，每次印出第幾次執行

```python
# 基本結構
counter = 0
while counter < 5:
    print(f"第 {counter + 1} 次執行")
    counter += 1
```

### For 迴圈 - range() 用法

prompt: 用 for 迴圈與 range(5) 印出 0~4，每次顯示數字

```python
# range() 用法
for i in range(5):  # 0, 1, 2, 3, 4
    print(f"數字: {i}")
```

### For 迴圈 - 遍歷清單

prompt: 用 for 迴圈遍歷一個水果清單，印出每個水果

```python
# 遍歷清單
fruits = ["蘋果", "香蕉", "橘子"]
for fruit in fruits:
    print(f"我喜歡吃{fruit}")
```

### 迴圈控制 - break 範例

prompt: 用 for 迴圈與 break，找到數字5時印出訊息並停止，否則印出數字

```python
# break 範例：找到目標就停止
for num in range(1, 11):
    if num == 5:
        print("找到 5 了！")
        break
    print(num)
```

### 迴圈控制 - continue 範例

prompt: 用 for 迴圈與 continue，跳過偶數只印出奇數

```python
# continue 範例：跳過偶數
for num in range(1, 6):
    if num % 2 == 0:
        continue
    print(f"奇數: {num}")
```

## 📋 清單 (List) 複習

### List 基本操作

prompt: 建立一個清單，示範新增、插入、刪除、修改與印出內容

```python
# 建立List
inventory = ["劍", "盾", "藥水"]

# 新增項目
inventory.append("弓箭")        # 加在最後
inventory.insert(1, "頭盔")     # 插入指定位置
print(inventory)

# 刪除項目
inventory.remove("藥水")        # 刪除指定項目
last_item = inventory.pop()    # 取出最後一個
print(inventory)
print(last_item)

# 修改項目
inventory[0] = "神劍"

print(inventory)
```

### List 常用方法實用範例

prompt: 建立一個分數清單，印出最高分、平均分、90分以上、排序前後

```python
# 實用範例
scores = [85, 92, 78, 95, 88]

print(f"最高分: {max(scores)}")
print(f"平均分: {sum(scores) / len(scores)}")
print(f"90分以上: {[s for s in scores if s >= 90]}")

print(f"排序前: {scores}")
scores.sort()
print(f"排序後: {scores}")
```

### 清單切片：靈活取值

prompt: 示範清單切片的多種用法，包含取部分、間隔、反轉等

```python
items = ["A", "B", "C", "D", "E", "F"]

print(items[1:4])      # ['B', 'C', 'D']
print(items[:3])       # ['A', 'B', 'C']
print(items[3:])       # ['D', 'E', 'F']
print(items[::2])      # ['A', 'C', 'E']  每2個取1個
print(items[::-1])     # ['F', 'E', 'D', 'C', 'B', 'A']  反轉
```

## 🎮 練習 1：寶物管理系統（完整版）

prompt: 寫一個寶物管理系統，能顯示、移除寶物，並處理清單空或無效輸入

```python
treasures = ["金幣", "寶石", "魔法書", "鑽石", "古董"]

while True:
    print("\n=== 寶物清單 ===")
    for i, treasure in enumerate(treasures, 1):
        print(f"{i}. {treasure}")

    choice = input("\n請選擇操作 (1-移除, 0-離開): ")

    if choice == "0":
        print("感謝使用！")
        break
    elif choice == "1":
        if len(treasures) == 0:
            print("清單已空！")
            continue

        index = int(input(f"請輸入要移除的編號(1-{len(treasures)}): ")) - 1
        if 0 <= index < len(treasures):
            removed = treasures.pop(index)
            print(f"已移除: {removed}")
        else:
            print("無效的編號！")
    else:
        print("無效的選擇！")
```

### 字典基本操作

prompt: 建立一個玩家字典，示範存取、新增、修改、刪除鍵值

```python
# 建立字典
player = {
    "name": "勇者",
    "hp": 100,
    "mp": 50,
    "level": 5
}
print(player)

# 存取值
print(player["name"])           # 勇者
print(player.get("exp", 0))    # 0 (不存在時回傳預設值)

# 新增/修改
player["weapon"] = "神劍"       # 新增
player["hp"] = 120              # 修改

# 刪除
del player["mp"]                # 刪除 mp
gold = player.pop("gold", 0)   # 取出 gold (不存在回傳 0)

print(player)
```

### 遍歷字典

prompt: 用 for 迴圈分別遍歷字典的鍵、值、鍵值對

```python
# 遍歷字典
player = {"name": "勇者", "hp": 100, "level": 5}

# 遍歷鍵
for key in player.keys():
    print(key)
print("-----")

# 遍歷值
for value in player.values():
    print(value)
print("-----")

# 遍歷鍵值對
for key, value in player.items():
    print(f"{key}: {value}")
print("-----")

# 遍歷鍵值對
for item in player.items():
    print(f"{item}")
```

### 函式 vs 無函式比較

prompt: 比較沒用函式與有用函式計算傷害的差異，並示範呼叫

```python
# 沒有函式的程式碼：重複、難維護
damage1 = 50 * 1.5 - 10
damage2 = 40 * 1.5 - 10
damage3 = 60 * 1.5 - 10

# 使用函式的程式碼：簡潔、可重用
def calculate_damage(attack, defense):
    return attack * 1.5 - defense

damage1 = calculate_damage(50, 10)
damage2 = calculate_damage(40, 10)
damage3 = calculate_damage(60, 10)
print(damage1, damage2, damage3)
```

### 變數作用域：區域與全域

prompt: 示範區域變數、全域變數與 global 關鍵字的用法

```python
# 全域變數
game_title = "冒險王國"

def show_status():
    # 區域變數
    player_hp = 100
    print(f"遊戲: {game_title}")
    print(f"HP: {player_hp}")

show_status()
# print(player_hp)  # 錯誤！區域變數在外部無法存取

# 修改全域變數需要 global 關鍵字
level = 1

def level_up():
    global level
    level += 1
    print(f"升級到 Lv.{level}")

level_up()
```

## 🚀 練習 2：角色管理系統（完整版）

prompt: 寫一個角色管理系統，能新增角色、顯示角色、角色升級，並示範呼叫

```python
characters = []

def add_character(name, job, hp, mp):
    """新增角色到清單"""
    character = {
        "name": name,
        "job": job,
        "hp": hp,
        "mp": mp,
        "level": 1
    }
    characters.append(character)
    print(f"✅ 角色 {name} 已建立！")

def show_all_characters():
    """顯示所有角色資訊"""
    if not characters:
        print("目前沒有角色！")
        return

    print("\n=== 角色列表 ===")
    for i, char in enumerate(characters, 1):
        print(f"{i}. {char['name']} ({char['job']}) - Lv.{char['level']}")
        print(f"   HP: {char['hp']} | MP: {char['mp']}")

def level_up_character(index):
    """角色升級：提升屬性"""
    if 0 <= index < len(characters):
        char = characters[index]
        char["level"] += 1
        char["hp"] += 20
        char["mp"] += 10
        print(f"🎉 {char['name']} 升級到 Lv.{char['level']}！")
    else:
        print("❌ 無效的角色編號！")

# 測試程式碼
add_character("亞瑟", "戰士", 150, 30)
add_character("梅林", "法師", 80, 120)
add_character("蘿賓", "弓手", 100, 60)
show_all_characters()
level_up_character(0)
show_all_characters()
```
