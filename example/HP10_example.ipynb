## ğŸš€ ç·´ç¿’ 2ï¼šè§’è‰²ç®¡ç†ç³»çµ±ï¼ˆå®Œæ•´ç‰ˆï¼‰

```python
characters = []

def add_character(name, job, hp, mp):
    """æ–°å¢è§’è‰²åˆ°æ¸…å–®"""
    character = {
        "name": name,
        "job": job,
        "hp": hp,
        "mp": mp,
        "level": 1
    }
    characters.append(character)
    print(f"âœ… è§’è‰² {name} å·²å»ºç«‹ï¼")

def show_all_characters():
    """é¡¯ç¤ºæ‰€æœ‰è§’è‰²è³‡è¨Š"""
    if not characters:
        print("ç›®å‰æ²’æœ‰è§’è‰²ï¼")
        return

    print("\n=== è§’è‰²åˆ—è¡¨ ===")
    for i, char in enumerate(characters, 1):
        print(f"{i}. {char['name']} ({char['job']}) - Lv.{char['level']}")
        print(f"   HP: {char['hp']} | MP: {char['mp']}")

def level_up_character(index):
    """è§’è‰²å‡ç´šï¼šæå‡å±¬æ€§"""
    if 0 <= index < len(characters):
        char = characters[index]
        char["level"] += 1
        char["hp"] += 20
        char["mp"] += 10
        print(f"ğŸ‰ {char['name']} å‡ç´šåˆ° Lv.{char['level']}ï¼")
    else:
        print("âŒ ç„¡æ•ˆçš„è§’è‰²ç·¨è™Ÿï¼")

# æ¸¬è©¦ç¨‹å¼ç¢¼
add_character("äºç‘Ÿ", "æˆ°å£«", 150, 30)
add_character("æ¢…æ—", "æ³•å¸«", 80, 120)
add_character("è˜¿è³“", "å¼“æ‰‹", 100, 60)
show_all_characters()
level_up_character(0)
show_all_characters()
```
