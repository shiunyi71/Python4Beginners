## 綜合實戰練習：資訊管理系統

**Prompt**: "寫一個 Python 學生資訊管理系統，包含新增資料、成績計算、隨機點名等功能。"

```python
import random

print("=== 🎓 學生資訊管理系統 ===")
print("1. 新增學生資料")
print("2. 成績計算與分析")
print("3. 隨機點名系統")
print("4. 離開系統")

choice = input("請選擇功能(1-4)：")

if choice == "1":
    # 新增學生資料
    name = input("學生姓名：")
    age = int(input("學生年齡："))
    grade = input("就讀年級：")
    print(f"✅ 學生 {name} 資料輸入完成")
    print(f"年齡：{age} 歲，年級：{grade}")
elif choice == "2":
    # 成績計算
    chinese = float(input("國文成績："))
    english = float(input("英文成績："))
    math = float(input("數學成績："))
    total = chinese + english + math
    average = total / 3
    print(f"總分：{total}")
    print(f"平均：{average:.2f}")
    if average >= 90:
        print("🏆 優秀！")
    elif average >= 80:
        print("👍 良好！")
    elif average >= 70:
        print("✅ 及格！")
    else:
        print("💪 需要加油！")
elif choice == "3":
    # 隨機點名系統
    students = ["小明", "小華", "小美", "小強", "小芳"]
    selected = random.choice(students)
    print(f"🎲 隨機點名：{selected}")
elif choice == "4":
    print("👋 系統結束，感謝使用！")
else:
    print("⚠️ 請輸入有效選項(1-4)")
```

## 參考答案

```python
import random

name = None
chinese = None

while True:
  print("=== 🎓 學生資訊管理系統 ===")
  print("1. 新增學生資料")
  print("2. 成績計算與分析")
  print("3. 隨機點名系統")
  print("4. 查詢學生資料")
  print("5. 修改學生成績")
  print("6. 離開系統")

  choice = input("請選擇功能(1-4)：")

  if choice == "1":
      # 新增學生資料
      name = input("學生姓名：")
      age = int(input("學生年齡："))
      grade = input("就讀年級：")
      print(f"✅ 學生 {name} 資料輸入完成")
      print(f"年齡：{age} 歲，年級：{grade}")
  elif choice == "2":
      # 成績計算
      chinese = float(input("國文成績："))
      english = float(input("英文成績："))
      math = float(input("數學成績："))
      total = chinese + english + math
      average = total / 3
      print(f"總分：{total}")
      print(f"平均：{average:.2f}")
      if average >= 90:
          print("🏆 優秀！")
      elif average >= 80:
          print("👍 良好！")
      elif average >= 70:
          print("✅ 及格！")
      else:
          print("💪 需要加油！")
  elif choice == "3":
      # 隨機點名系統
      students = ["小明", "小華", "小美", "小強", "小芳"]
      selected = random.choice(students)
      print(f"🎲 隨機點名：{selected}")
  elif choice == "4":
    if name:
      # 查學生資料
      print(f"✅ 學生 {name} 資料輸入完成")
      print(f"年齡：{age} 歲，年級：{grade}")
    else:
      print("查無此人")
  elif choice == "5":
    if chinese:
      # 修改學生成績
      mod = input('請問要改哪科？1國/2英/3數')
      if mod == '1':
        chinese = float(input("國文成績："))
      elif mod == '2':
        english = float(input("英文成績："))
      elif mod == '3':
        math = float(input("數學成績："))
      total = chinese + english + math
      average = total / 3
      print(f"總分：{total}")
      print(f"平均：{average:.2f}")
    else:
      print("請先輸入成績後才能修改")


  elif choice == "6":
      print("👋 系統結束，感謝使用！")
      break
  else:
      print("⚠️ 請輸入有效選項(1-6)的數字")
```

## Tkinter 基本視窗

**Prompt**: "用 Tkinter 建立一個基本視窗，顯示 Hello, Tkinter!"

```python
import tkinter as tk

window = tk.Tk()
window.title("Tkinter 基本視窗")
window.geometry("300x200")

label = tk.Label(window, text="Hello, Tkinter!", font=("Arial", 16))
label.pack(pady=20)

window.mainloop()
```

## Tkinter 按鈕與事件

**Prompt**: "用 Tkinter 建立一個有按鈕的視窗，按下按鈕會顯示訊息。"

```python
import tkinter as tk

def say_hello():
    label.config(text="你按下了按鈕！")

window = tk.Tk()
window.title("按鈕範例")

label = tk.Label(window, text="請按下按鈕")
label.pack()

button = tk.Button(window, text="點我", command=say_hello)
button.pack()

window.mainloop()
```
