## 字典方法大全

```python
backpack = {"劍": 1, "盾": 1, "藥水": 5}

# 重要方法
keys = backpack.keys()      # 所有鍵
values = backpack.values()  # 所有值
items = backpack.items()    # 所有鍵值對
print(keys)
print(values)
print(items)

# 安全取值
# count = backpack["弓箭"]  # 不存在會壞掉
count = backpack.get("弓箭", 0)  # 不存在返回0
print(f"弓箭數量：{count}")

# 更新字典
new_items = {"弓箭": 1, "箭矢": 50}
backpack.update(new_items)
print(f"更新後背包內容：{backpack}")
```

## 字典的增刪查改範例

```python
# 增加新鍵值對
backpack["飛箭"] = 2
print(backpack)  # {'劍': 1, '治療藥水': 5, '魔法書': 2, '飛箭': 2}

# 查詢某個鍵的值
print("劍", backpack.get("劍"))  # 2
print("魔法書", backpack.get("魔法書"))  # 2
print("盾牌", backpack.get("盾牌", "沒有此物品"))  # 沒有此物品

# 修改現有鍵的值
backpack["藥水"] = 10
print("藥水", backpack["藥水"])  # 10

# 刪除鍵值對
del backpack["長劍"] #長劍不存在會錯
print(backpack)  # {'治療藥水': 10, '魔法書': 2, '弓箭': 2}

# 使用 pop() 刪除並取得值
removed = backpack.pop("弓箭", None)
print(f"移除弓箭，數量：{removed}")  # 移除弓箭，數量：2
print(backpack)  # {'藥水': 10, '魔法書': 2}
```

## RPG背包管理系統（互動式選單）

```python
backpack = {}

while True:
  print("\n=== 背包管理系統 ===")
  print("1. 新增道具及數量")
  print("2. 查詢某個道具的數量")
  print("3. 修改道具的數量")
  print("4. 刪除某個道具")
  print("5. 顯示所有道具")
  print("0. 離開")
  choice = input("請輸入選項：")

  if choice == "1":
    name = input("請輸入道具名稱：")
    qty = int(input("請輸入數量："))
    if name in backpack:
      backpack[name] += qty
    else:
      backpack[name] = qty
    print(f"已新增 {qty} 個 {name}")
  elif choice == "2":
    name = input("請輸入要查詢的道具名稱：")
    print(f"{name} 數量：{backpack.get(name, 0)}")

  elif choice == "3":
    name = input("請輸入要修改的道具名稱：")
    if name in backpack:
      qty = int(input("請輸入新的數量："))
      backpack[name] = qty
      print(f"{name} 數量已修改為 {qty}")
    else:
      print("背包中沒有此道具")

  elif choice == "4":
    name = input("請輸入要刪除的道具名稱：")
    if name in backpack:
      del backpack[name]
      print(f"{name} 已刪除")
    else:
      print("背包中沒有此道具")

  elif choice == "5":
    print("=== 背包內容 ===")
    for item, qty in backpack.items():
      print(f"{item}: {qty}")

  elif choice == "0":
    print("離開背包管理系統")
    break

  else:
    print("請輸入有效選項")
```

# 補充資料 - Set（集合）
## 範例：集合運算

```python
s = {1, 2, 3, 2}
print(s)        # {1, 2, 3}
s.add(4)        # 新增元素
s.remove(2)     # 移除元素
print(s)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)    # 聯集 {1, 2, 3, 4, 5}
print(a & b)    # 交集 {3}
print(a - b)    # 差集 {1, 2}
print(a ^ b)    # 互異 {1, 2, 4, 5}
```

```python
# 超集合與子集合
superset_a = {1, 2, 3, 4, 5}
subset_b = {2, 3}
subset_c = {6, 7}

print(f"集合 A: {superset_a}")
print(f"集合 B: {subset_b}")
print(f"集合 C: {subset_c}")
print(f"B 是 A 的子集合嗎？ {subset_b.issubset(superset_a)}")  # True
print(f"A 是 B 的超集合嗎？ {superset_a.issuperset(subset_b)}")  # True
print(f"C 是 A 的子集合嗎？ {subset_c.issubset(superset_a)}")  # False
print(f"A 是 C 的超集合嗎？ {superset_a.issuperset(subset_c)}")  # False
```

```python

print(f"B 是 A 的子集合嗎？ {subset_b <= superset_a}")  # True
print(f"A 是 B 的超集合嗎？ {superset_a >= subset_b}")  # True
print(f"C 是 A 的子集合嗎？ {subset_c <= superset_a}")  # False
print(f"A 是 C 的超集合嗎？ {superset_a >= subset_c}")  # False
```
