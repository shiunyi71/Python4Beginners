```python
# 超簡單動畫：跑馬燈效果（來回）
import time
text = 'Python 好好玩！'
width = 10
for i in list(range(width)) + list(range(width - 2, 0, -1)):
    print('\r' + ' ' * i + text, end='')
    time.sleep(0.2)
print()
```
