```python
import random

def generate_random_treasure():
    """éš¨æ©Ÿç”Ÿæˆå¯¶ç‰©"""
    treasure_names = [
        "ç¥ç§˜å¯¶ç®±", "å¤è€å·è»¸", "é­”æ³•æ°´æ™¶", "é¾é±—ç›¾ç‰Œ",
        "å‚³èªªä¹‹åŠ", "æ²»ç™‚è—¥æ°´", "éš±èº«æ–—ç¯·", "ç«ç„°æˆ’æŒ‡"
    ]

    rarities = ["æ™®é€š", "ç¨€æœ‰", "å²è©©", "å‚³èªª"]
    rarity_multipliers = {"æ™®é€š": 1, "ç¨€æœ‰": 3, "å²è©©": 10, "å‚³èªª": 50}

    name = random.choice(treasure_names)
    rarity = random.choice(rarities)
    base_value = random.randint(10, 100)
    final_value = base_value * rarity_multipliers[rarity]

    return {
        "name": f"{rarity}{name}",
        "value": final_value,
        "rarity": rarity
    }

# ç”Ÿæˆ5å€‹éš¨æ©Ÿå¯¶ç‰©
for _ in range(5):
    treasure = generate_random_treasure()
    print(f"{treasure['name']} - åƒ¹å€¼: {treasure['value']}")
```

```python
def linear_search(treasure_list, target):
    """ç·šæ€§æœå°‹å¯¶ç‰©"""
    for i, treasure in enumerate(treasure_list):
        if treasure == target:
            return i  # å›å‚³ç´¢å¼•ä½ç½®
    return -1  # æ‰¾ä¸åˆ°

treasures = ["é‡‘å¹£", "éŠ€åŠ", "é­”æ³•æ›¸", "ç´…å¯¶çŸ³"]
position = linear_search(treasures, "é­”æ³•æ›¸")
if position != -1:
    print(f"æ‰¾åˆ°é­”æ³•æ›¸åœ¨ä½ç½® {position}")
else:
    print("æ‰¾ä¸åˆ°é­”æ³•æ›¸")
```

## ç¶œåˆç¯„ä¾‹ï¼šå¯¶ç‰©åˆ†é¡å™¨

```python
# æ‰€æœ‰å¯¶ç‰©ï¼ˆæœªåˆ†é¡ï¼‰
all_items = ["é‡‘åŠ", "éŠ€å¹£", "éŠ€åŠ", "é­”æ³•æ›¸", "é‡‘å¹£", "éµç›¾", "ç´…è—¥æ°´", "é‡‘æˆ’æŒ‡", "è—è—¥æ°´"]

# å»ºç«‹åˆ†é¡æ¸…å–®
weapons = []      # æ­¦å™¨
money = []        # é‡‘éŒ¢
books = []        # æ›¸ç±
potions = []      # è—¥æ°´
jewelry = []      # é£¾å“

# è«‹å®Œæˆåˆ†é¡é‚è¼¯
for item in all_items:
    if "åŠ" in item or "ç›¾" in item:
        weapons.append(item)
    elif "å¹£" in item:
        money.append(item)
    elif "æ›¸" in item:
        books.append(item)
    elif "è—¥æ°´" in item:
        potions.append(item)
    elif "æˆ’æŒ‡" in item:
        jewelry.append(item)

print(f"æ­¦å™¨: {weapons}")
print(f"é‡‘éŒ¢: {money}")
print(f"æ›¸ç±: {books}")
print(f"è—¥æ°´: {potions}")
print(f"é£¾å“: {jewelry}")
```

```python
def main():
  treasures = []
  total_value = 0

  while True:
    print("\nğŸ° === å¯¶ç‰©ç®¡ç†ç³»çµ± ===")
    print("1. æ·»åŠ å¯¶ç‰©")
    print("2. ç§»é™¤å¯¶ç‰©")
    print("3. æœå°‹å¯¶ç‰©")
    print("4. é¡¯ç¤ºæ¸…å–®")
    print("5. æœ€æœ‰åƒ¹å€¼å¯¶ç‰©")
    print("6. é›¢é–‹")

    choice = input("\nè«‹é¸æ“‡æ“ä½œ (1-6): ")
    if choice == "1":
      name = input("å¯¶ç‰©åç¨±: ")
      value = int(input("å¯¶ç‰©åƒ¹å€¼: "))
      treasure = {"name": name, "value": value}
      treasures.append(treasure)
      total_value += value
      print(f"âœ… æˆåŠŸæ·»åŠ å¯¶ç‰©: {name} (åƒ¹å€¼: {value})")

    elif choice == "2":
      name = input("è¦ç§»é™¤çš„å¯¶ç‰©åç¨±: ")
      found = False
      for treasure in treasures:
        if treasure["name"] == name:
          treasures.remove(treasure)
          total_value -= treasure["value"]
          print(f"ğŸ—‘ï¸ æˆåŠŸç§»é™¤å¯¶ç‰©: {name}")
          found = True
          break
      if not found:
        print(f"âŒ æ‰¾ä¸åˆ°å¯¶ç‰©: {name}")

    elif choice == "3":
      keyword = input("æœå°‹é—œéµå­—: ")
      results = [treasure for treasure in treasures if keyword.lower() in treasure["name"].lower()]
      if results:
        print("\nğŸ” æœå°‹çµæœ:")
        for treasure in results:
          print(f"- {treasure['name']} (åƒ¹å€¼: {treasure['value']})")
      else:
        print("âŒ æ‰¾ä¸åˆ°ç›¸é—œå¯¶ç‰©")

    elif choice == "4":
      print("\nğŸ’ === å¯¶ç‰©æ¸…å–® ===")
      if not treasures:
        print("èƒŒåŒ…æ˜¯ç©ºçš„ï¼")
      else:
        for i, treasure in enumerate(treasures, 1):
          print(f"{i}. {treasure['name']} - åƒ¹å€¼: {treasure['value']}")
        print(f"\nğŸ’° ç¸½åƒ¹å€¼: {total_value}")

    elif choice == "5":
      if not treasures:
        print("èƒŒåŒ…æ˜¯ç©ºçš„ï¼")
      else:
        most_valuable = max(treasures, key=lambda x: x["value"])
        print(f"\nğŸ’ æœ€æœ‰åƒ¹å€¼çš„å¯¶ç‰©: {most_valuable['name']} (åƒ¹å€¼: {most_valuable['value']})")

    elif choice == "6":
      print("ğŸ‘‹ å†è¦‹ï¼")
      break

    else:
      print("âŒ ç„¡æ•ˆçš„é¸æ“‡ï¼Œè«‹é‡æ–°è¼¸å…¥ï¼")

# åŸ·è¡Œç¨‹å¼
if __name__ == "__main__":
  main()
```
