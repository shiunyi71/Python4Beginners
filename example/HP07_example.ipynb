## 隨機寶物生成器

### Prompt 範例
請用 Python 撰寫一個隨機生成寶物的函式，並用 for 迴圈產生多個隨機寶物並印出。

```python
import random

def generate_random_treasure():
    """隨機生成寶物"""
    treasure_names = [
        "神秘寶箱", "古老卷軸", "魔法水晶", "龍鱗盾牌",
        "傳說之劍", "治療藥水", "隱身斗篷", "火焰戒指"
    ]

    rarities = ["普通", "稀有", "史詩", "傳說"]
    rarity_multipliers = {"普通": 1, "稀有": 3, "史詩": 10, "傳說": 50}

    name = random.choice(treasure_names)
    rarity = random.choice(rarities)
    base_value = random.randint(10, 100)
    final_value = base_value * rarity_multipliers[rarity]

    return {
        "name": f"{rarity}{name}",
        "value": final_value,
        "rarity": rarity
    }

# 生成5個隨機寶物
for _ in range(5):
    treasure = generate_random_treasure()
    print(f"{treasure['name']} - 價值: {treasure['value']}")
```

### Prompt 範例
請用 Python 撰寫一個線性搜尋函式，搜尋清單中指定寶物的位置，並示範如何使用。

```python
def linear_search(treasure_list, target):
    """線性搜尋寶物"""
    for i, treasure in enumerate(treasure_list):
        if treasure == target:
            return i  # 回傳索引位置
    return -1  # 找不到

treasures = ["金幣", "銀劍", "魔法書", "紅寶石"]
position = linear_search(treasures, "魔法書")
if position != -1:
    print(f"找到魔法書在位置 {position}")
else:
    print("找不到魔法書")
```

### Prompt 範例
請用 Python 寫一個 for 迴圈線性搜尋範例，找出清單中指定寶物的位置，若找不到則印出提示。

```python
treasures = ["金幣", "銀劍", "魔法書", "紅寶石"]

"""線性搜尋寶物"""
target = "銀劍"
for i, treasure in enumerate(treasures):
    if treasure == target:
        print(f"找到{target}在位置 {i}")
        break
else:
    print("找不到"+target)
```

## 綜合範例：寶物分類器

### Prompt 範例
請用 Python 將一個寶物清單依照類型分類到不同清單，並印出分類結果。

```python
# 所有寶物（未分類）
all_items = ["金劍", "銀幣", "銀劍", "魔法書", "金幣", "鐵盾", "紅藥水", "金戒指", "藍藥水"]

# 建立分類清單
weapons = []      # 武器
money = []        # 金錢
books = []        # 書籍
potions = []      # 藥水
jewelry = []      # 飾品

# 請完成分類邏輯
for item in all_items:
    if "劍" in item or "盾" in item:
        weapons.append(item)
    elif "幣" in item:
        money.append(item)
    elif "書" in item:
        books.append(item)
    elif "藥水" in item:
        potions.append(item)
    elif "戒指" in item:
        jewelry.append(item)

print(f"武器: {weapons}")
print(f"金錢: {money}")
print(f"書籍: {books}")
print(f"藥水: {potions}")
print(f"飾品: {jewelry}")
```

## 實作專案：寶物收集管理系統

### Prompt 範例
請用 Python 設計一個互動式寶物收集管理系統，能新增、移除、搜尋、顯示與找出最有價值寶物。

```python
def main():
  treasures = []
  total_value = 0

  while True:
    print("\n🏰 === 寶物管理系統 ===")
    print("1. 添加寶物")
    print("2. 移除寶物")
    print("3. 搜尋寶物")
    print("4. 顯示清單")
    print("5. 最有價值寶物")
    print("6. 離開")

    choice = input("\n請選擇操作 (1-6): ")
    if choice == "1":
      name = input("寶物名稱: ")
      value = int(input("寶物價值: "))
      treasure = {"name": name, "value": value}
      treasures.append(treasure)
      total_value += value
      print(f"✅ 成功添加寶物: {name} (價值: {value})")

    elif choice == "2":
      name = input("要移除的寶物名稱: ")
      found = False
      for treasure in treasures:
        if treasure["name"] == name:
          treasures.remove(treasure)
          total_value -= treasure["value"]
          print(f"🗑️ 成功移除寶物: {name}")
          found = True
          break
      if not found:
        print(f"❌ 找不到寶物: {name}")

    elif choice == "3":
      keyword = input("搜尋關鍵字: ")
      results = [treasure for treasure in treasures if keyword.lower() in treasure["name"].lower()]
      if results:
        print("\n🔍 搜尋結果:")
        for treasure in results:
          print(f"- {treasure['name']} (價值: {treasure['value']})")
      else:
        print("❌ 找不到相關寶物")

    elif choice == "4":
      print("\n🎒 === 寶物清單 ===")
      if not treasures:
        print("背包是空的！")
      else:
        for i, treasure in enumerate(treasures, 1):
          print(f"{i}. {treasure['name']} - 價值: {treasure['value']}")
        print(f"\n💰 總價值: {total_value}")

    elif choice == "5":
      if not treasures:
        print("背包是空的！")
      else:
        most_valuable = max(treasures, key=lambda x: x["value"])
        print(f"\n💎 最有價值的寶物: {most_valuable['name']} (價值: {most_valuable['value']})")

    elif choice == "6":
      print("👋 再見！")
      break

    else:
      print("❌ 無效的選擇，請重新輸入！")

# 執行程式
if __name__ == "__main__":
  main()
```
