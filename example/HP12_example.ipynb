# 用迴圈讀取所有行

prompt: 示範用 while 迴圈和 for 迴圈兩種方式讀取檔案所有行

```python
print("=== 用 while 迴圈讀取所有行 ===")
with open('my_diary.txt', 'r', encoding='utf-8') as file:
    line_number = 1
    while True:
        line = file.readline()
        if not line:  # 讀到檔案結尾
            break
        print(f'第 {line_number} 行: {line.strip()}')
        line_number += 1

print("\n=== 更簡單的方法：直接迴圈檔案物件 ===")
with open('my_diary.txt', 'r', encoding='utf-8') as file:
    for line_number, line in enumerate(file, 1):
        print(f'第 {line_number} 行: {line.strip()}')
```

## 實際練習：處理學生成績檔案

prompt: 建立學生成績 CSV 檔案並讀取計算各科平均和班級比較

```python
# 建立學生成績檔案
student_scores = """小明,85,92,78
小華,90,88,95
小美,87,91,89
小強,76,82,84
小麗,94,96,91"""

with open('student_scores.csv', 'w', encoding='utf-8') as f:
    f.write("姓名,國文,英文,數學\n")  # CSV標題行
    f.write(student_scores)

print("✅ 學生成績檔案已建立")

# 讀取並處理成績資料
print("\n=== 讀取成績資料 ===")
with open('student_scores.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# 處理資料
print("成績統計：")
header = lines[0].strip().split(',')
print(f"欄位: {header}")
print()

total_scores = [0, 0, 0]  # 三科總分
student_count = 0

for line in lines[1:]:  # 跳過標題行
    parts = line.strip().split(',')
    name = parts[0]
    scores = [int(parts[i]) for i in range(1, 4)]
    average = sum(scores) / len(scores)

    print(f"{name}: {scores} (平均: {average:.1f})")

    # 累加各科分數
    for i in range(3):
        total_scores[i] += scores[i]
    student_count += 1

# 計算各科平均
print("\n各科班級平均：")
subjects = ['國文', '英文', '數學']
for i in range(3):
    avg = total_scores[i] / student_count
    print(f"{subjects[i]}: {avg:.1f}")
```

# 實用技巧：自動加換行

prompt: 示範三種自動加換行符號的方法：串列生成式、join()、迴圈

```python
print("=== 自動加換行的方法 ===")

# 原始資料（沒有換行符號）
lines_data = ['第一行', '第二行', '第三行', '第四行']

# 方法1：用串列生成式自動加上 \n
print("方法1：串列生成式")
with open('method1.txt', 'w', encoding='utf-8') as file:
    file.writelines([line + '\n' for line in lines_data])

with open('method1.txt', 'r', encoding='utf-8') as file:
    print("結果1：", repr(file.read()))
print()

# 方法2：用 join() 方法
print("方法2：join() 方法")
with open('method2.txt', 'w', encoding='utf-8') as file:
    file.write('\n'.join(lines_data) + '\n')

with open('method2.txt', 'r', encoding='utf-8') as file:
    print("結果2：", repr(file.read()))
print()

# 方法3：用迴圈逐一寫入
print("方法3：迴圈寫入")
with open('method3.txt', 'w', encoding='utf-8') as file:
    for line in lines_data:
        file.write(line + '\n')

with open('method3.txt', 'r', encoding='utf-8') as file:
    print("結果3：", repr(file.read()))
```

# 格式化寫入範例

prompt: 示範使用 f-string 格式化寫入學生資料到檔案

```python
print("=== 格式化寫入範例 ===")

# 學生資料
students = [
    {'name': '小明', 'age': 15, 'score': 95.5},
    {'name': '小華', 'age': 16, 'score': 87.2},
    {'name': '小美', 'age': 15, 'score': 92.8}
]

# 寫入格式化的學生資料
with open('students.txt', 'w', encoding='utf-8') as file:
    file.write("=== 學生成績單 ===\n")
    file.write(f"{'姓名':^6} {'年齡':^6} {'成績':^8}\n")
    file.write("-" * 25 + "\n")

    for student in students:
        file.write(f"{student['name']:^6} {student['age']:^6} {student['score']:^8.1f}\n")

print("✅ 學生資料已寫入")

# 讀取並顯示
with open('students.txt', 'r', encoding='utf-8') as file:
    content = file.read()
    print("檔案內容：")
    print(content)
```

# 進階範例：自動日記系統

prompt: 建立自動日記系統函式，自動加上時間戳記和表情符號

```python
from datetime import datetime

def write_diary_entry(content, mood="😊"):
    """寫入一筆日記，自動加上時間戳記"""
    now = datetime.now()
    date_str = now.strftime('%Y-%m-%d %H:%M:%S')
    weekday = ['一', '二', '三', '四', '五', '六', '日'][now.weekday()]

    with open('auto_diary.txt', 'a', encoding='utf-8') as file:
        file.write(f'[{date_str} 星期{weekday}] {mood}\n')
        file.write(f'{content}\n')
        file.write('-' * 50 + '\n')

    print(f"✅ 日記已記錄：{content[:20]}...")

# 使用範例
write_diary_entry('今天學會了檔案處理，感覺很有成就感！', '🎉')
write_diary_entry('練習了讀取和寫入檔案的各種方法', '📚')
write_diary_entry('明天想繼續學習更進階的內容', '💪')

# 讀取日記內容
print("\n=== 我的自動日記 ===")
try:
    with open('auto_diary.txt', 'r', encoding='utf-8') as file:
        diary_content = file.read()
        print(diary_content)
except FileNotFoundError:
    print("還沒有日記記錄")
```

# 實用範例：處理 CSV 格式

prompt: 建立學生成績 CSV 檔案並進行班級比較和科目分析

```python
print("=== CSV 檔案處理範例 ===")

# 學生資料（模擬真實情況）
students_data = [
    ['姓名', '班級', '國文', '英文', '數學'],
    ['王小明', '1A', '85', '92', '78'],
    ['李小華', '1B', '90', '88', '95'],
    ['張小美', '1A', '87', '91', '89'],
    ['陳小強', '1B', '76', '82', '84'],
    ['林小麗', '1A', '94', '96', '91']
]

# 寫入 CSV 檔案
print("建立學生成績 CSV 檔案...")
with open('students_grades.csv', 'w', encoding='utf-8') as file:
    for row in students_data:
        file.write(','.join(row) + '\n')

print("✅ CSV 檔案已建立")

# 讀取並分析 CSV 檔案
print("\n=== 分析成績資料 ===")
with open('students_grades.csv', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 處理標題行
header = lines[0].strip().split(',')
print(f"欄位：{header}")
print()

# 分析每個學生的成績
class_scores = {'1A': [], '1B': []}
all_chinese = []
all_english = []
all_math = []

for line in lines[1:]:
    data = line.strip().split(',')
    name, class_name = data[0], data[1]
    chinese, english, math = int(data[2]), int(data[3]), int(data[4])
    total = chinese + english + math
    average = total / 3

    print(f"{name} ({class_name}): 國文{chinese}, 英文{english}, 數學{math} "
          f"→ 總分{total}, 平均{average:.1f}")

    # 收集統計資料
    class_scores[class_name].append(average)
    all_chinese.append(chinese)
    all_english.append(english)
    all_math.append(math)

# 班級比較
print("\n=== 班級平均比較 ===")
for class_name in ['1A', '1B']:
    class_avg = sum(class_scores[class_name]) / len(class_scores[class_name])
    print(f"{class_name}班平均：{class_avg:.1f}")

# 科目分析
print("\n=== 各科平均分數 ===")
subjects = [('國文', all_chinese), ('英文', all_english), ('數學', all_math)]
for subject_name, scores in subjects:
    avg = sum(scores) / len(scores)
    highest = max(scores)
    lowest = min(scores)
    print(f"{subject_name}: 平均{avg:.1f}, 最高{highest}, 最低{lowest}")
```

# 進階應用：讀取修改再寫入

prompt: 讀取日記檔案、加上行號和裝飾後寫入新的格式化檔案

```python
print("=== 讀取、修改、寫回檔案 ===")

# 讀取原有的日記內容
try:
    with open('my_diary.txt', 'r', encoding='utf-8') as file:
        lines = file.readlines()

    print("原始日記內容：")
    for i, line in enumerate(lines, 1):
        print(f"{i}. {line.strip()}")

    # 修改內容：加上行號和日期格式化
    modified_lines = []
    for i, line in enumerate(lines, 1):
        if line.strip():  # 不是空行
            # 添加行號和裝飾
            modified_line = f"📝 第{i}筆記錄: {line.strip()}\n"
            modified_lines.append(modified_line)

    # 寫回檔案
    with open('my_diary_formatted.txt', 'w', encoding='utf-8') as file:
        file.write("=" * 50 + "\n")
        file.write("📔 我的格式化日記本\n")
        file.write("=" * 50 + "\n\n")
        file.writelines(modified_lines)
        file.write("\n" + "=" * 50 + "\n")
        file.write(f"📊 總共有 {len(modified_lines)} 筆記錄\n")

    print("\n✅ 格式化日記已建立")

    # 顯示修改後的內容
    print("\n修改後的內容：")
    with open('my_diary_formatted.txt', 'r', encoding='utf-8') as file:
        formatted_content = file.read()
        print(formatted_content)

except FileNotFoundError:
    print("找不到原始日記檔案，請先執行前面的範例建立日記檔案")
```

### 練習1：個人記帳本
建立一個簡單的記帳程式，可以：
1. 記錄收入和支出
2. 計算餘額
3. 顯示交易記錄

```python
# 練習：個人記帳本
print("=== 個人記帳本系統 ===")

def add_transaction(date, description, amount, type_="支出"):
    """新增交易記錄"""
    with open('account_book.txt', 'a', encoding='utf-8') as file:
        sign = "+" if type_ == "收入" else "-"
        file.write(f"{date},{description},{sign}{amount},{type_}\n")
    print(f"✅ 已記錄{type_}：{description} {sign}{amount}元")

def show_transactions():
    """顯示所有交易記錄"""
    try:
        with open('account_book.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        if not lines:
            print("還沒有任何交易記錄")
            return

        print("\n=== 交易記錄 ===")
        total = 0
        for i, line in enumerate(lines, 1):
            parts = line.strip().split(',')
            date, desc, amount_str, trans_type = parts
            amount = int(amount_str)
            total += amount

            print(f"{i:2d}. {date} | {desc:10s} | {amount_str:>6s}元 | {trans_type}")

        print("-" * 50)
        print(f"💰 目前餘額：{total:+d}元")

    except FileNotFoundError:
        print("還沒有交易記錄檔案")

# 範例交易
add_transaction("2024-01-15", "零用錢", 1000, "收入")
add_transaction("2024-01-15", "午餐", 80, "支出")
add_transaction("2024-01-15", "文具", 150, "支出")
add_transaction("2024-01-16", "打工", 500, "收入")
add_transaction("2024-01-16", "交通費", 60, "支出")

# 顯示記錄
show_transactions()
```

### 練習2：簡易通訊錄
建立一個通訊錄系統，可以：
1. 新增聯絡人
2. 搜尋聯絡人  
3. 顯示所有聯絡人

```python
# 練習：簡易通訊錄
print("=== 簡易通訊錄系統 ===")

def add_contact(name, phone, email=""):
    """新增聯絡人"""
    with open('contacts.txt', 'a', encoding='utf-8') as file:
        file.write(f"{name},{phone},{email}\n")
    print(f"✅ 已新增聯絡人：{name}")

def search_contact(keyword):
    """搜尋聯絡人"""
    try:
        with open('contacts.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        found = []
        for line in lines:
            parts = line.strip().split(',')
            name, phone, email = parts
            if keyword.lower() in name.lower() or keyword in phone:
                found.append((name, phone, email))

        if found:
            print(f"\n🔍 搜尋到 {len(found)} 筆資料：")
            for name, phone, email in found:
                print(f"   📱 {name}: {phone} ({email})")
        else:
            print(f"❌ 找不到包含 '{keyword}' 的聯絡人")

    except FileNotFoundError:
        print("❌ 通訊錄檔案不存在")

def show_all_contacts():
    """顯示所有聯絡人"""
    try:
        with open('contacts.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        if not lines:
            print("📱 通訊錄是空的")
            return

        print(f"\n📞 通訊錄 (共{len(lines)}筆)")
        print("-" * 40)
        for i, line in enumerate(lines, 1):
            parts = line.strip().split(',')
            name, phone, email = parts
            email_display = f"({email})" if email else ""
            print(f"{i:2d}. {name:8s} | {phone:12s} {email_display}")

    except FileNotFoundError:
        print("❌ 通訊錄檔案不存在")

# 範例聯絡人
add_contact("小明", "0912-345-678", "ming@email.com")
add_contact("小華", "0923-456-789", "hua@email.com")
add_contact("小美", "0934-567-890")
add_contact("老師", "02-1234-5678", "teacher@school.edu")

# 測試功能
show_all_contacts()
search_contact("小")
search_contact("02")
search_contact("老師")
```

```python
# filepath: diary_system.py
from datetime import datetime

def write_diary():
    """寫入新的日記"""
    print('\n=== 寫日記 ===')
    content = input('請輸入今天的日記內容：')

    now = datetime.now()
    date_str = now.strftime('%Y-%m-%d %H:%M:%S')

    with open('my_diary.txt', 'a', encoding='utf-8') as file:
        file.write('=' * 40 + '\n')
        file.write(f'日期：{date_str}\n')
        file.write('=' * 40 + '\n')
        file.write(f'{content}\n\n')

    print('✅ 日記已保存！')

# filepath: diary_system.py
def read_diary():
    """閱讀所有日記"""
    print('\n=== 我的日記本 ===')
    try:
        with open('my_diary.txt', 'r', encoding='utf-8') as file:
            content = file.read()
            if content:
                print(content)
            else:
                print('日記本是空的，快來寫第一篇日記吧！')
    except FileNotFoundError:
        print('還沒有日記檔案，請先寫一篇日記！')

# filepath: diary_system.py
def search_diary():
    """搜尋日記"""
    print('\n=== 搜尋日記 ===')
    keyword = input('請輸入要搜尋的關鍵字：')

    try:
        with open('my_diary.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()
            found = False
            current_entry = []

            for line in lines:
                current_entry.append(line)
                if line.strip() == '' and current_entry:
                    entry_text = ''.join(current_entry)
                    if keyword in entry_text:
                        print(entry_text)
                        found = True
                    current_entry = []

            if not found:
                print(f'沒有找到包含「{keyword}」的日記')
    except FileNotFoundError:
        print('還沒有日記檔案！')

# filepath: diary_system.py
def clear_diary():
    """清空所有日記"""
    print('\n=== 清空日記 ===')
    confirm = input('確定要刪除所有日記嗎？(yes/no)：')
    if confirm.lower() == 'yes':
        with open('my_diary.txt', 'w', encoding='utf-8') as file:
            pass  # 開啟 'w' 模式但不寫入，檔案就被清空了
        print('✅ 日記已清空')
    else:
        print('❌ 取消清空')


# filepath: diary_system.py
def show_menu():
    """顯示選單"""
    print('\n' + '=' * 40)
    print('📔 歡迎使用個人日記系統')
    print('=' * 40)
    print('1. ✍  寫新日記')
    print('2. 📖 閱讀所有日記')
    print('3. 🔍 搜尋日記')
    print('4. 🗑  清空日記')
    print('5. 👋 離開系統')
    print('=' * 40)

# filepath: diary_system.py
def main():
    """主程式"""
    while True:
        show_menu()
        choice = input('請選擇功能 (1-5)：')

        if choice == '1':
            write_diary()
        elif choice == '2':
            read_diary()
        elif choice == '3':
            search_diary()
        elif choice == '4':
            clear_diary()
        elif choice == '5':
            print('👋 謝謝使用，再見！')
            break
        else:
            print('❌ 無效的選擇，請輸入 1-5')

if __name__ == '__main__':
    main()
```
