## å¯¦éš›ç·´ç¿’ï¼šè™•ç†å­¸ç”Ÿæˆç¸¾æª”æ¡ˆ

```python
# å»ºç«‹å­¸ç”Ÿæˆç¸¾æª”æ¡ˆ
student_scores = """å°æ˜,85,92,78
å°è¯,90,88,95
å°ç¾,87,91,89
å°å¼·,76,82,84
å°éº—,94,96,91"""

with open('student_scores.csv', 'w', encoding='utf-8') as f:
    f.write("å§“å,åœ‹æ–‡,è‹±æ–‡,æ•¸å­¸\n")  # CSVæ¨™é¡Œè¡Œ
    f.write(student_scores)

print("âœ… å­¸ç”Ÿæˆç¸¾æª”æ¡ˆå·²å»ºç«‹")

# è®€å–ä¸¦è™•ç†æˆç¸¾è³‡æ–™
print("\n=== è®€å–æˆç¸¾è³‡æ–™ ===")
with open('student_scores.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# è™•ç†è³‡æ–™
print("æˆç¸¾çµ±è¨ˆï¼š")
header = lines[0].strip().split(',')
print(f"æ¬„ä½: {header}")
print()

total_scores = [0, 0, 0]  # ä¸‰ç§‘ç¸½åˆ†
student_count = 0

for line in lines[1:]:  # è·³éæ¨™é¡Œè¡Œ
    parts = line.strip().split(',')
    name = parts[0]
    scores = [int(parts[i]) for i in range(1, 4)]
    average = sum(scores) / len(scores)

    print(f"{name}: {scores} (å¹³å‡: {average:.1f})")

    # ç´¯åŠ å„ç§‘åˆ†æ•¸
    for i in range(3):
        total_scores[i] += scores[i]
    student_count += 1

# è¨ˆç®—å„ç§‘å¹³å‡
print("\nå„ç§‘ç­ç´šå¹³å‡ï¼š")
subjects = ['åœ‹æ–‡', 'è‹±æ–‡', 'æ•¸å­¸']
for i in range(3):
    avg = total_scores[i] / student_count
    print(f"{subjects[i]}: {avg:.1f}")
```

# å¯¦ç”¨ç¯„ä¾‹ï¼šè™•ç† CSV æ ¼å¼

```python
print("=== CSV æª”æ¡ˆè™•ç†ç¯„ä¾‹ ===")

# å­¸ç”Ÿè³‡æ–™ï¼ˆæ¨¡æ“¬çœŸå¯¦æƒ…æ³ï¼‰
students_data = [
    ['å§“å', 'ç­ç´š', 'åœ‹æ–‡', 'è‹±æ–‡', 'æ•¸å­¸'],
    ['ç‹å°æ˜', '1A', '85', '92', '78'],
    ['æå°è¯', '1B', '90', '88', '95'],
    ['å¼µå°ç¾', '1A', '87', '91', '89'],
    ['é™³å°å¼·', '1B', '76', '82', '84'],
    ['æ—å°éº—', '1A', '94', '96', '91']
]

# å¯«å…¥ CSV æª”æ¡ˆ
print("å»ºç«‹å­¸ç”Ÿæˆç¸¾ CSV æª”æ¡ˆ...")
with open('students_grades.csv', 'w', encoding='utf-8') as file:
    for row in students_data:
        file.write(','.join(row) + '\n')

print("âœ… CSV æª”æ¡ˆå·²å»ºç«‹")

# è®€å–ä¸¦åˆ†æ CSV æª”æ¡ˆ
print("\n=== åˆ†ææˆç¸¾è³‡æ–™ ===")
with open('students_grades.csv', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# è™•ç†æ¨™é¡Œè¡Œ
header = lines[0].strip().split(',')
print(f"æ¬„ä½ï¼š{header}")
print()

# åˆ†ææ¯å€‹å­¸ç”Ÿçš„æˆç¸¾
class_scores = {'1A': [], '1B': []}
all_chinese = []
all_english = []
all_math = []

for line in lines[1:]:
    data = line.strip().split(',')
    name, class_name = data[0], data[1]
    chinese, english, math = int(data[2]), int(data[3]), int(data[4])
    total = chinese + english + math
    average = total / 3

    print(f"{name} ({class_name}): åœ‹æ–‡{chinese}, è‹±æ–‡{english}, æ•¸å­¸{math} "
          f"â†’ ç¸½åˆ†{total}, å¹³å‡{average:.1f}")

    # æ”¶é›†çµ±è¨ˆè³‡æ–™
    class_scores[class_name].append(average)
    all_chinese.append(chinese)
    all_english.append(english)
    all_math.append(math)

# ç­ç´šæ¯”è¼ƒ
print("\n=== ç­ç´šå¹³å‡æ¯”è¼ƒ ===")
for class_name in ['1A', '1B']:
    class_avg = sum(class_scores[class_name]) / len(class_scores[class_name])
    print(f"{class_name}ç­å¹³å‡ï¼š{class_avg:.1f}")

# ç§‘ç›®åˆ†æ
print("\n=== å„ç§‘å¹³å‡åˆ†æ•¸ ===")
subjects = [('åœ‹æ–‡', all_chinese), ('è‹±æ–‡', all_english), ('æ•¸å­¸', all_math)]
for subject_name, scores in subjects:
    avg = sum(scores) / len(scores)
    highest = max(scores)
    lowest = min(scores)
    print(f"{subject_name}: å¹³å‡{avg:.1f}, æœ€é«˜{highest}, æœ€ä½{lowest}")
```

### ç·´ç¿’1ï¼šå€‹äººè¨˜å¸³æœ¬

```python
# ç·´ç¿’ï¼šå€‹äººè¨˜å¸³æœ¬
print("=== å€‹äººè¨˜å¸³æœ¬ç³»çµ± ===")

def add_transaction(date, description, amount, type_="æ”¯å‡º"):
    """æ–°å¢äº¤æ˜“è¨˜éŒ„"""
    with open('account_book.txt', 'a', encoding='utf-8') as file:
        sign = "+" if type_ == "æ”¶å…¥" else "-"
        file.write(f"{date},{description},{sign}{amount},{type_}\n")
    print(f"âœ… å·²è¨˜éŒ„{type_}ï¼š{description} {sign}{amount}å…ƒ")

def show_transactions():
    """é¡¯ç¤ºæ‰€æœ‰äº¤æ˜“è¨˜éŒ„"""
    try:
        with open('account_book.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        if not lines:
            print("é‚„æ²’æœ‰ä»»ä½•äº¤æ˜“è¨˜éŒ„")
            return

        print("\n=== äº¤æ˜“è¨˜éŒ„ ===")
        total = 0
        for i, line in enumerate(lines, 1):
            parts = line.strip().split(',')
            date, desc, amount_str, trans_type = parts
            amount = int(amount_str)
            total += amount

            print(f"{i:2d}. {date} | {desc:10s} | {amount_str:>6s}å…ƒ | {trans_type}")

        print("-" * 50)
        print(f"ğŸ’° ç›®å‰é¤˜é¡ï¼š{total:+d}å…ƒ")

    except FileNotFoundError:
        print("é‚„æ²’æœ‰äº¤æ˜“è¨˜éŒ„æª”æ¡ˆ")

# ç¯„ä¾‹äº¤æ˜“
add_transaction("2024-01-15", "é›¶ç”¨éŒ¢", 1000, "æ”¶å…¥")
add_transaction("2024-01-15", "åˆé¤", 80, "æ”¯å‡º")
add_transaction("2024-01-15", "æ–‡å…·", 150, "æ”¯å‡º")
add_transaction("2024-01-16", "æ‰“å·¥", 500, "æ”¶å…¥")
add_transaction("2024-01-16", "äº¤é€šè²»", 60, "æ”¯å‡º")

# é¡¯ç¤ºè¨˜éŒ„
show_transactions()
```

```python
# ç·´ç¿’ï¼šç°¡æ˜“é€šè¨ŠéŒ„
print("=== ç°¡æ˜“é€šè¨ŠéŒ„ç³»çµ± ===")

def add_contact(name, phone, email=""):
    """æ–°å¢è¯çµ¡äºº"""
    with open('contacts.txt', 'a', encoding='utf-8') as file:
        file.write(f"{name},{phone},{email}\n")
    print(f"âœ… å·²æ–°å¢è¯çµ¡äººï¼š{name}")

def search_contact(keyword):
    """æœå°‹è¯çµ¡äºº"""
    try:
        with open('contacts.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        found = []
        for line in lines:
            parts = line.strip().split(',')
            name, phone, email = parts
            if keyword.lower() in name.lower() or keyword in phone:
                found.append((name, phone, email))

        if found:
            print(f"\nğŸ” æœå°‹åˆ° {len(found)} ç­†è³‡æ–™ï¼š")
            for name, phone, email in found:
                print(f"   ğŸ“± {name}: {phone} ({email})")
        else:
            print(f"âŒ æ‰¾ä¸åˆ°åŒ…å« '{keyword}' çš„è¯çµ¡äºº")

    except FileNotFoundError:
        print("âŒ é€šè¨ŠéŒ„æª”æ¡ˆä¸å­˜åœ¨")

def show_all_contacts():
    """é¡¯ç¤ºæ‰€æœ‰è¯çµ¡äºº"""
    try:
        with open('contacts.txt', 'r', encoding='utf-8') as file:
            lines = file.readlines()

        if not lines:
            print("ğŸ“± é€šè¨ŠéŒ„æ˜¯ç©ºçš„")
            return

        print(f"\nğŸ“ é€šè¨ŠéŒ„ (å…±{len(lines)}ç­†)")
        print("-" * 40)
        for i, line in enumerate(lines, 1):
            parts = line.strip().split(',')
            name, phone, email = parts
            email_display = f"({email})" if email else ""
            print(f"{i:2d}. {name:8s} | {phone:12s} {email_display}")

    except FileNotFoundError:
        print("âŒ é€šè¨ŠéŒ„æª”æ¡ˆä¸å­˜åœ¨")

# ç¯„ä¾‹è¯çµ¡äºº
add_contact("å°æ˜", "0912-345-678", "ming@email.com")
add_contact("å°è¯", "0923-456-789", "hua@email.com")
add_contact("å°ç¾", "0934-567-890")
add_contact("è€å¸«", "02-1234-5678", "teacher@school.edu")

# æ¸¬è©¦åŠŸèƒ½
show_all_contacts()
search_contact("å°")
search_contact("02")
search_contact("è€å¸«")
```
